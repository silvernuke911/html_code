```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

def generate_pixel_combinations():
    return list(itertools.product([0, 1], repeat=9))

def display_combinations(batch_size=10):
    combinations = generate_pixel_combinations()
    
    for i in range(0, len(combinations), batch_size):
        batch = combinations[i:i+batch_size]
        n = len(batch)
        cols = 5
        rows = (n + cols - 1) // cols

        fig, axes = plt.subplots(rows, cols, figsize=(cols * 2, rows * 2))
        axes = axes.flatten()

        for ax, comb in zip(axes, batch):
            grid = np.array(comb).reshape((3, 3))
            ax.imshow(grid, cmap='gray', vmin=0, vmax=1)
            ax.set_title(f"{i + axes.tolist().index(ax) + 1}")
            ax.axis('off')

        # Hide unused subplots
        for j in range(n, len(axes)):
            axes[j].axis('off')

        plt.tight_layout()
        plt.show()

display_combinations(batch_size=10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_counter_image(n, invert = True):
    # Determine how many bits are needed to represent n - 1
    num_bits = n.bit_length()
    
    # Create a binary matrix: each row is the binary of a number from 0 to n-1
    binary_matrix = np.array([
        [int(b) for b in f"{i:0{num_bits}b}"]
        for i in range(n)
    ])

    if invert:
        binary_matrix = 1 - binary_matrix

    plt.figure(figsize = (5,10), dpi = 300)
    plt.imshow(binary_matrix, cmap='gray', vmin=0, vmax=1)
    plt.title(f'Binary counter up to {n-1}')
    plt.xlabel('Bits (MSB  LSB)')
    plt.ylabel('Decimal value')
    plt.xticks(ticks=np.arange(num_bits), labels=[f'B{i}' for i in range(num_bits)])
    plt.yticks(ticks=np.arange(0,n,10))#, labels=[str(i) for i in range(n)])
    plt.grid(False)
    plt.show()

# Example usage:
binary_counter_image(256)
```
